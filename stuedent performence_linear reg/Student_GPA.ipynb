## <div class="custom-section"><div class="custom-header">step 2.1 = data cleaning</div><div class="custom-content">define cleaning function</div></div>



```python
# Preprocessing step
def data_clean(x):
    # Fill missing values (median) for 'ParentalEducation' and 'ParentalSupport'
    x["ParentalEducation"].fillna(x["ParentalEducation"].mode()[0], inplace=True)
    x["ParentalSupport"].fillna(x["ParentalSupport"].mode()[0], inplace=True)

    # One-hot encode 'Ethnicity' and concatenate with the original DataFrame
    ethnicity_dummies = pd.get_dummies(x["Ethnicity"])
    x = pd.concat([x, ethnicity_dummies], axis=1)
    x.drop(columns="Ethnicity", axis=1, inplace=True)
    x[['African American', 'Asian' , 'Caucasian', 'Other']] = x[['African American', 'Asian' , 'Caucasian', 'Other']].astype("int64")

    # Map categorical values to numerical (using replace=True if needed)
    education_mapping = {"High School": 1, "Some College": 2, "Bachelor": 3, "Higher": 4}
    x["ParentalEducation"] = x["ParentalEducation"].map(education_mapping)

    support_mapping = {"Low": 1, "Moderate": 2, "High": 3, "Very High": 4}
    x["ParentalSupport"] = x["ParentalSupport"].map(support_mapping)

    x.dropna(inplace=True)

    return x

# Apply the function
df = data_clean(df)
test_df = data_clean(test_df)
df
```

## <div class="custom-section"><div class="custom-header">Step 2.2 | split and scale data</div><div class="custom-content">train_test_split & StandardScaler</div></div>


```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=["StudentID", "GPA"])
y = df["GPA"]
X_test = test_df.drop(columns=['StudentID'])

X_train, X_val, y_train, y_val = train_test_split(X,y, random_state=42, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numerical_features = ['Age', 'StudyTimeWeekly', 'Absences']

X_train[numerical_features] = scaler.fit_transform(X_train[numerical_features])
X_val[numerical_features] = scaler.transform(X_val[numerical_features])
X_test[numerical_features] = scaler.transform(X_test[numerical_features])
```
