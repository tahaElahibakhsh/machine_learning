```python
# 8
a = np.random.randint(low=-100, high=100, size=(3,4))

min_val = np.min(a, axis=1, keepdims=True)
max_val = np.max(a, axis=1, keepdims=True)

a_norm = (a - min_val) / (max_val - min_val)

print(np.mean(a_norm, axis=1))
print(np.std(a_norm, axis=1))
print(a_norm)
```
