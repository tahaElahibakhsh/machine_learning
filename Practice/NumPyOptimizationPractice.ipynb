```python
a = np.random.randint(low=1, high=11, size=(3,2))

# with for:
norm_a = np.zeros(a.shape)
for i in range(a.shape[0]):
    s = 0
    for j in range(a.shape[1]):
        s += a[i, j]
    for j in range(a.shape[1]):
        norm_a[i, j] = a[i, j] / s
    
print(a)
print(norm_a)


# with numpy array:
row_sum = np.sum(a, axis=1, keepdims=True)  # shape (3, 1)
norm_a_numpy = a / row_sum
print(norm_a_numpy)
```

```python
a = np.random.randint(low=1, high=11, size=(3,2))

# with for:
out = np.zeros(a.shape)
for j in range(a.shape[1]):
    mean_j = 0
    for i in range(a.shape[0]):
        mean_j += a[i, j]
    mean_j /= a.shape[0]
    
    for i in range(a.shape[0]):
        out[i, j] = a[i, j] - mean_j
    
print(a)
print(out)


# with numpy array:
col_sum = np.mean(a, axis=0, keepdims=True)
norm_a_numpy = a - col_sum
print(norm_a_numpy)
```

```python
X = np.random.randint(low=1, high=11, size=(4,5))

# with for:
D = np.zeros((X.shape[0], X.shape[0]))
for i in range(X.shape[0]):
    for k in range(X.shape[0]):
        s = 0
        for j in range(X.shape[1]):
            diff = X[i, j] - X[k, j]
            s += diff ** 2
        D[i, k] = s
        
print(X)
print(D)


# with numpy array:
row_norms = np.sum(X ** 2, axis=1)
dot_product = X @ X.T
D_formula = row_norms[:, np.newaxis] + row_norms[np.newaxis, :] - 2 * dot_product

print(D_formula)
```

```python
Z = np.random.randn(2, 3)

# with for:
out = np.zeros(Z.shape)
for i in range(Z.shape[0]):
    denom = 0
    for j in range(Z.shape[1]):
        denom += np.exp(Z[i, j])
    for j in range(Z.shape[1]):
        out[i, j] = np.exp(Z[i, j]) / denom
        
        
print(Z)
print(out)


# with numpy
max_vals = np.max(Z, axis=1, keepdims=True)
Z_shifted = Z - max_vals
exp_Z = np.exp(Z_shifted)
sum_exp = np.sum(exp_Z, axis=1, keepdims=True) 
softmax = exp_Z / sum_exp


print(softmax)
```

```python
n_samples = 5
n_features = 3

X = np.random.randint(low=0, high=11, size=(n_samples, n_features))

centered = X - np.mean(X, axis=0, keepdims=True)

cov = (centered.T @ centered) / (n_samples - 1)

std = np.std(X, axis=0, ddof=1, keepdims=True)
standardized = centered / std
corr = (standardized.T @ standardized) / (n_samples - 1)

print(X)
print(cov)
print(corr)


cov = np.cov(X, rowvar=False)
corr = np.corrcoef(X, rowvar=False)
print(cov)
print(corr)
```

```python
m, n = 10, 2
k = 3

X = np.random.randn(m, n)
C = np.random.randn(k, n)

def kmeans_assignment(X, C):
    X_norm = np.sum(X**2, axis=1, keepdims=True)
    C_norm = np.sum(C**2, axis=1, keepdims=True).T
    XC = X @ C.T
    distances = X_norm + C_norm - 2 * XC
    labels = np.argmin(distances, axis=1)
    return labels, distances

labels, distances = kmeans_assignment(X, C)

print(labels, distances)
```
