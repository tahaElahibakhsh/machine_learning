## <div class="custom-section"><div class="custom-header">Step 2.1 | fit model</div><div class="custom-content">k_means.fit(data)</div></div>



```python
k_means = KMeans(n_clusters=6, random_state=42)
k_means.fit(data)
centroids = k_means.cluster_centers_
labels = k_means.labels_


print("cluster centers :")
for i, center in enumerate(centroids):
    print(f"cluster {i}: ({center[0]:.2f}, {center[1]:.2f})")
```

## <div class="custom-section"><div class="custom-header">Step 2.2 | visualize</div><div class="custom-content">scatter plot of data</div></div>



```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y, c=labels, cmap='viridis', s=10, alpha=0.7)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='*', s=120, label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.grid(True)
plt.show()
```
