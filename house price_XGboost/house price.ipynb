## <div class="custom-section"><div class="custom-header">step 2.1 = data cleaning</div><div class="custom-content">define cleaning function</div></div>

## <div class="custom-section"><div class="custom-header">Step 2.4 | add feature</div><div class="custom-content">add HouseAge & RemodAge & TotalBath & TotalPorchSF</div></div>


```python
current_year = pd.Timestamp.now().year
df_train['HouseAge'] = current_year - df_train['YearBuilt']
df_train['RemodAge'] = current_year - df_train['YearRemodAdd']


# Create total bathrooms feature
df_train['TotalBath'] = df_train['FullBath'] + (0.5 * df_train['HalfBath']) + \
                  df_train['BsmtFullBath'] + (0.5 * df_train['BsmtHalfBath'])

# Create total porch area
df_train['TotalPorchSF'] = df_train['OpenPorchSF'] + df_train['EnclosedPorch'] + \
                     df_train['3SsnPorch'] + df_train['ScreenPorch']
```

## <div class="custom-section"><div class="custom-header">Step 2.5 | encoding</div><div class="custom-content">encode categorical data with label and onnehot encoder</div></div>


```python
cat = df_train.select_dtypes(include=["category", "object"]).columns

for col in cat:
    unique_count = df_train[col].nunique()
    print(f"col '{col}': {unique_count}")
```

```python
from sklearn.preprocessing import LabelEncoder

# Columns to one-hot encode
one_hot_cols = ['MSZoning', 'Street', 'LotShape', 'LandContour', 'LotConfig', 
               'LandSlope', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl',
               'Foundation', 'Heating', 'CentralAir', 'Electrical', 'Functional',
               'GarageType', 'PavedDrive', 'SaleType', 'SaleCondition','Neighborhood','Condition2','Condition1']

# Columns to label encode
label_cols = ['ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond', 'BsmtExposure',
             'BsmtFinType1', 'BsmtFinType2', 'HeatingQC', 'KitchenQual',
             'GarageFinish', 'GarageQual', 'GarageCond','Exterior2nd','Exterior1st']



# Apply one-hot encoding
df_train = pd.get_dummies(df_train, columns=one_hot_cols, drop_first=True)

# Apply label encoding
le = LabelEncoder()
for col in label_cols:
    df_train[col] = le.fit_transform(df_train[col])
    
```

```python
bool_cols = df_train.select_dtypes(include=['bool']).columns
df_train[bool_cols] = df_train[bool_cols].astype(int)
df_train.info()
```

## <div class="custom-section"><div class="custom-header">Step 3.3 | metrics</div><div class="custom-content">RMSE & R²score</div></div>


```python
y_train_pred = model.predict(x_train)
y_val_pred = model.predict(x_val)



mse_train = mean_squared_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)
rmse_train = np.sqrt(mse_train)


mse_val = mean_squared_error(y_val, y_val_pred)
r2_val = r2_score(y_val, y_val_pred)
rmse_val = np.sqrt(mse_val)


print(f"RMSE: {rmse_train:.2f}")
print("-----------------------------------------------")
print(f"MSE for validation: {mse_val:.2f}")
print(f"R² for validation: {r2_val:.2f}")
print(f"RMSE: {rmse_val:.2f}")
```
