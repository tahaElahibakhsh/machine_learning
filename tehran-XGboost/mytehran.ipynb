## <div class="custom-section"><div class="custom-header">step 2.1 = data cleaning</div><div class="custom-content">define cleaning function</div></div>



```python
data['Area'] = data['Area'].str.replace(" ", "")
data['Area'] = data['Area'].str.replace(",", "")
```

```python
data['Elevator'] = data['Elevator'].astype("int")
data['Parking'] = data['Parking'].astype("int")
data['Area'] = data['Area'].astype("int64")
data['Warehouse'] = data['Warehouse'].astype("int64")
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

counts = data["Address"].value_counts()

rare_labels = counts[counts < 8].index

data["Address"] = data["Address"].replace(rare_labels, "other")

le = LabelEncoder()
data["Address"] = le.fit_transform(data["Address"])

data["Address"].value_counts()
```

```python
data['Area'].plot(kind='box')
plt.show()
```

```python
data['Area_log'] = np.log(data['Area'])
```

```python
Q1 = data.Area.quantile(0.25)
Q2 = data.Area.quantile(0.50)
Q3 = data.Area.quantile(0.75)
IQR = Q3 - Q1
LC = Q1 - (1.5*IQR)
UC = Q3 + (1.5*IQR)
display(LC)
display(UC)
```

```python
data = data[(data['Area'] >= LC) & (data['Area'] <= UC)].reset_index(drop=True)
```

## <div class="custom-section"><div class="custom-header">Step 2.4 | add feature</div><div class="custom-content">add HouseAge & RemodAge & TotalBath & TotalPorchSF</div></div>


```python
def estimate_floors(row):
    area = row["Area"]
    rooms = row["Room"]
    elevator = row["Elevator"]  # 0 یا 1
    
    if rooms == 0:
        return 1

    ratio = area / rooms  # نسبت مساحت به اتاق

    # منطق پایه
    if ratio < 25:
        floor = 1
    elif ratio < 55:
        floor = 1
    elif ratio < 90:
        floor = 2
    else:
        floor = 3

    # تاثیر آسانسور
    if elevator == 0 and floor > 2:
        floor = 2  # اگر آسانسور نیست، حداکثر 2 طبقه
    elif elevator == 1 and floor == 3:
        floor = 3  # با آسانسور اجازه طبقه بالاتر

    return floor

data["FloorCount"] = data.apply(estimate_floors, axis=1)
```

```python
data["AreaPerRoom"] = data["Area"] / data["Room"].replace(0, 1)
data["AreaPerFloor"] = data["Area"] / data["FloorCount"].replace(0, 1)
data["ParkingPerRoom"] = data["Parking"] / data["Room"].replace(0,1)
data["FloorElevator"] = data["FloorCount"] * data["Elevator"]
data["AreaFloorElevator"] = data["AreaPerFloor"] * data["Elevator"]
data["LargeArea"] = (data["Area"] > data["Area"].median()).astype(int)
data["LuxuryUnit"] = ((data["Parking"]>0) & (data["Elevator"]==1) & (data["Room"]>=3)).astype(int)
data["PricePerArea"] = data["Price"] / data["Area"].replace(0,1)
data["PricePerRoom"] = data["Price"] / data["Room"].replace(0,1)
freq = data["Address"].value_counts()
data["AddressFreq"] = data["Address"].map(freq)
```

```python
g = data.groupby("Address")["Price(USD)"]

data["addr_price_mean"] = data["Address"].map(g.mean())
data["addr_price_min"]  = data["Address"].map(g.min())
data["addr_price_max"]  = data["Address"].map(g.max())
```

## <div class="custom-section"><div class="custom-header">Step 3.1 | split data</div><div class="custom-content">train_test_split</div></div>


```python
from sklearn.model_selection import train_test_split

# Sample data
X = data.drop(columns=['Price(USD)', "Price"])  # Features
y = data['Price']  # Target variable

# Split the data: 80% training, 20% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Training set:", X_train.shape, y_train.shape)
print("Testing set:", X_test.shape, y_test.shape)
```

```python
from sklearn.preprocessing import MaxAbsScaler, MinMaxScaler, StandardScaler, RobustScaler, Normalizer, Binarizer

ss = StandardScaler()
X_train_scale = ss.fit_transform(X_train)
```

```python
X_test_scale = ss.transform(X_test)
```

## <div class="custom-section"><div class="custom-header">Step 3.2 | fit model</div><div class="custom-content">model = XGBRegressor()</div></div>


```python
from xgboost import XGBRegressor
from sklearn.model_selection import RandomizedSearchCV

# محدوده‌های پارامترها
param_dist = {
    "n_estimators": [1200, 1300, 1500, 1800, 2000],
    "learning_rate": [0.01, 0.02, 0.03, 0.05, 0.08],
    "max_depth": [7, 4, 5, 6],
    "min_child_weight": [6, 6, 3, 4, 5],
    "subsample": [0.5, 0.6, 0.7, 0.8],
    "colsample_bytree": [0.8, 0.5, 0.6, 0.7],
    "gamma": [0, 0.5, 0.2, 0.3, 0.4],
    "reg_alpha": [0.8, 0.3, 0.5, 0.7, 1.0],
    "reg_lambda": [1, 1.25, 1.5, 1.75, 2.0]
}

# مدل پایه
xgb = XGBRegressor(
    tree_method="hist",
    random_state=56
)

# RandomizedSearchCV
rs = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_dist,
    n_iter=50,           # تعداد ترکیب‌های تصادفی
    scoring="neg_mean_absolute_error",
    cv=3,
    verbose=1,
    n_jobs=-1,
    random_state=56
)

# اجرای تیونینگ
rs.fit(X_train_scale, y_train)

# بهترین مدل
best_model = rs.best_estimator_

# پیش‌بینی
y_pred = best_model.predict(X_test_scale)
y_pred_t = best_model.predict(X_train_scale)

# بهترین پارامترها
print("Best Params:", rs.best_params_)
```
